**Задача 6**\
Сгенерировать матрицу B размером 5 на 5, элементы которой лежат в интервале [-5, 5]. Отрицательные элементы матрицы заменить на число отрицательных элементов в строке, в которой они находятся.

```python
b = np.random.randint(-5,6, size = (5,5))
b
```

```python
for i in range(b.shape[0]):  # Проходим по каждой строке матрицы
    neg_count = np.sum(b[i] < 0)  # Считаем количество отрицательных элементов в текущей строке
    b[i, b[i] < 0] = neg_count
```

# Вывод:
Некоторые переменные имеют значимые связи между собой, что может указывать на их взаимосвязь или общие факторы, влияющие на них.
SAT и Acceptance: Наблюдается сильная отрицательная корреляция между результатами тестов SAT и процентом принятых абитуриентов (-0.61). Это может означать, что школы с более высокими результатами SAT имеют более низкий процент принятых абитуриентов, и наоборот. Это может быть связано с более высокими требованиями к поступлению в учебные заведения с более высокими результатами SAT.

SAT и $/Student: Наблюдается средняя положительная корреляция между результатами тестов SAT и средними расходами на одного студента (0.58). Это может указывать на то, что школы с более высокими результатами SAT имеют более высокие средние расходы на одного студента.

Acceptance и Top 10%: Наблюдается сильная отрицательная корреляция между процентом принятых абитуриентов и процентом студентов в топ 10% выпускных классов (-0.62). Это может означать, что школы, которые принимают меньше студентов, чаще имеют более качественный пул абитуриентов, что проявляется в более высоком проценте студентов в топ 10%.

Top 10% и PhD%: Наблюдается средняя положительная корреляция между процентом студентов в топ 10% выпускных классов и процентом факультетов, имеющих PhD (0.45). Это может указывать на то, что школы, где большой процент студентов входит в топ 10%, также имеют факультет с более высокой академической квалификацией.

Grad% и SAT: Наблюдается средняя положительная корреляция между процентом успешных завершений обучения и результатами тестов SAT (0.57). Это может означать, что школы с более высокими результатами SAT также имеют более высокий процент успешных завершений обучения.

**Задание 4**\
Найдите:
+ выборочное среднее SAT школ искусств;
+ 3й выборочный квартиль Acceptance университетов;
+ название института(колледжа), в котором тратят денег в среднем на одного студента больше, чем в других институтах(колледжах).

```python
# Выборочное среднее SAT школ искусств
mean_sat_lib_arts = df[df['School_Type'] == 'Lib Arts']['SAT'].mean()
print("Выборочное среднее SAT школ искусств:", mean_sat_lib_arts)
```

```python
# 3-й выборочный квартиль Acceptance университетов
q3 = df[df['School_Type'] == 'Univ']['Acceptance'].quantile(0.75)
print("3-й выборочный квартиль Acceptance университетов:", q3)
```

```python
max_spending_college = df[df['$/Student'] == df['$/Student'].max()]['School'].values[0]
print("Институт (колледж), в котором тратят денег в среднем на одного студента больше:", max_spending_college)
```
